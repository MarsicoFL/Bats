```python
import pandas as pd
import numpy as np

def custom_compare(ind1, ind2):
    # Filtrando los NA
    non_na_indices = ~(np.isnan(ind1) | np.isnan(ind2))
    filtered_ind1 = ind1[non_na_indices]
    filtered_ind2 = ind2[non_na_indices]

    # Aplicando la lógica de comparación personalizada
    coincidences = ((filtered_ind1 == 0) & (filtered_ind2 == 0)) | \
                   ((filtered_ind1 == 2) & (filtered_ind2 == 2)) | \
                   ((filtered_ind1 == 0) & (filtered_ind2 == 1)) | \
                   ((filtered_ind1 == 2) & (filtered_ind2 == 1))
    discrepancies = ((filtered_ind1 == 0) & (filtered_ind2 == 2)) | \
                    ((filtered_ind1 == 2) & (filtered_ind2 == 0))

    return np.sum(coincidences), np.sum(discrepancies), len(filtered_ind1)

def main():
    # Lectura del archivo CSV
    file_path = '/content/drive/MyDrive/Genomics/Bats/chrX.csv'
    data = pd.read_csv(file_path)

    # Eliminando las columnas que no son de genotipos
    genotype_data = data.drop(['MARKER_ID', 'CHR', 'Mb_v0423', 'REF', 'ALT'], axis=1)

    # Filtrando los marcadores donde uno de los individuos tiene valor 1
    filtro = genotype_data['GSO-129-b'] != 1
    filtered_ind1 = genotype_data['GSO-129-b'][filtro]
    filtered_ind2 = genotype_data['GSO-84-m'][filtro]

    # Realizando la comparación personalizada
    matches, discrepancies, total_comparisons = custom_compare(filtered_ind1, filtered_ind2)

    # Imprimiendo los resultados
    print(f'Coincidencias: {matches}')
    print(f'Discrepancias: {discrepancies}')
    print(f'Total de comparaciones: {total_comparisons}')

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np

# Función para la comparación personalizada
def custom_compare(ind1, ind2):
    coincidences = ((ind1 == 0) & (ind2 == 0)) | \
                   ((ind1 == 2) & (ind2 == 2)) | \
                   ((ind1 == 0) & (ind2 == 1)) | \
                   ((ind1 == 2) & (ind2 == 1))
    discrepancies = ((ind1 == 0) & (ind2 == 2)) | \
                    ((ind1 == 2) & (ind2 == 0))
    return np.sum(coincidences), np.sum(discrepancies), len(ind1)

# Función para realizar la comparación considerando el sexo
def custom_compare_with_sex(ind1_id, ind2_id, genotype_data, sex_dict):
    ind1 = genotype_data[ind1_id]
    ind2 = genotype_data[ind2_id]

    # Aplicar filtro de valor 1 si alguno de los individuos es masculino
    if sex_dict.get(ind1_id, None) == 'M' or sex_dict.get(ind2_id, None) == 'M':
        filtro = (ind1 != 1) & (ind2 != 1)
    else:
        filtro = np.ones(len(ind1), dtype=bool)  # No aplicar filtro si ambos son F

    filtered_ind1 = ind1[filtro]
    filtered_ind2 = ind2[filtro]

    return custom_compare(filtered_ind1, filtered_ind2)

# Lectura de los archivos CSV
genotype_data = pd.read_csv('/content/drive/MyDrive/Genomics/Bats/chrX.csv').drop(['MARKER_ID', 'CHR', 'Mb_v0423', 'REF', 'ALT'], axis=1)
metadata = pd.read_csv('/content/drive/MyDrive/Genomics/Bats/BatMetadata.csv', sep='\t')
sex_dict = dict(zip(metadata['Animal_ID'], metadata['Sex']))

# Realizando comparaciones
comparison_results = {}
for individual in genotype_data.columns:
    if individual != 'GSO-129-b':
        matches, discrepancies, total_comparisons = custom_compare_with_sex('GSO-129-b', individual, genotype_data, sex_dict)
        comparison_results[individual] = {
            'coincidencias': matches,
            'discrepancias': discrepancies,
            'total_comparaciones': total_comparisons
        }

# Calculando porcentajes de coincidencias
comparison_df = pd.DataFrame.from_dict(comparison_results, orient='index')
comparison_df['porcentaje_coincidencias'] = (comparison_df['coincidencias'] / comparison_df['total_comparaciones']) * 100

# Guardando los resultados
comparison_df.to_csv('resultados_comparacion.csv')

# Identificando pares con más del 95% de coincidencias
pares_alta_coincidencia = comparison_df[comparison_df['porcentaje_coincidencias'] > 10]
pares_alta_coincidencia.to_csv('pares_alta_coincidencia.csv')
```
